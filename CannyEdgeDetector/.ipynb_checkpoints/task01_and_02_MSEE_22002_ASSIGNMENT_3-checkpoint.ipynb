```python
def findFilterXYDerivative(filter_size, sigma):
    sHalf = round((-2*math.log(T)*(sigma**2))**(1/2))
    print(sHalf)
    N= 2*sHalf+1
    return np.meshgrid (np.linspace(-sHalf, sHalf, N), np.linspace(-sHalf , sHalf,N))

def findFlip2D(imgFilt):
    output=np.zeros([imgFilt.shape[0],imgFilt.shape[1]])
    for y in range(0,imgFilt.shape[0]):
        for x in range(0,imgFilt.shape[1]):
            output[y,x]=imgFilt[imgFilt.shape[0]-y-1,imgFilt.shape[1]-x-1]
    return output

def findGaussian(N, sigma):
    x, y = np.mgrid[-N//2 + 1:N//2 + 1, -N//2 + 1:N//2 + 1]
    g = np.exp(-((x**2 + y**2)/(2.0*sigma**2)))
    return g/g.sum()

def findDotProduct(imgFilt,imgPatch):
    if(imgFilt.shape==imgPatch.shape):
        return sum(sum(imgFilt*imgPatch))
    else:
        print("Dimensions of given inputs do not match.")
        print(imgFilt.shape)
        print(imgPatch.shape)
        return
    
def findConvolution2D(imgFiltr,imgTarget):
    
    if (imgTarget.shape[0]-imgFiltr.shape[0]>-1)and(imgTarget.shape[1]-imgFiltr.shape[1]>-1):
        imgFilt=findFlip2D(imgFiltr)
        output=np.zeros([imgTarget.shape[0]+imgFilt.shape[0]-1,imgTarget.shape[1]+imgFilt.shape[1]-1])
        temp=np.zeros([imgTarget.shape[0]+2*imgFilt.shape[0]-2,imgTarget.shape[1]+2*imgFilt.shape[1]-2])
        temp[imgFilt.shape[0]-1:imgTarget.shape[0]+imgFilt.shape[0]-1,imgFilt.shape[1]-1:imgTarget.shape[1]+imgFilt.shape[1]-1]=imgTarget
        for y in range(0,output.shape[0]):
            for x in range(0,output.shape[1]):
                output[y,x]=findDotProduct(imgFilt,temp[y:y+imgFilt.shape[0],x:x+imgFilt.shape[1]])
        return output
    else:
        print("Target image dimension is less than the target image dimension")
        return
    
def findConvolution2Dfixed(imgFiltr,imgTarget):
    
    if (imgTarget.shape[0]-imgFiltr.shape[0]>-1)and(imgTarget.shape[1]-imgFiltr.shape[1]>-1):
        imgFilt=findFlip2D(imgFiltr)
        output=np.zeros([imgTarget.shape[0]+imgFilt.shape[0]-1,imgTarget.shape[1]+imgFilt.shape[1]-1])
        temp=np.zeros([imgTarget.shape[0]+2*imgFilt.shape[0]-2,imgTarget.shape[1]+2*imgFilt.shape[1]-2])
        temp[imgFilt.shape[0]-1:imgTarget.shape[0]+imgFilt.shape[0]-1,imgFilt.shape[1]-1:imgTarget.shape[1]+imgFilt.shape[1]-1]=imgTarget
        for y in range(0,output.shape[0]):
            for x in range(0,output.shape[1]):
                output[y,x]=findDotProduct(imgFilt,temp[y:y+imgFilt.shape[0],x:x+imgFilt.shape[1]])
        return output[imgFilt.shape[0]:output.shape[0]-imgFilt.shape[0]+1,imgFilt.shape[1]:output.shape[1]-imgFilt.shape[1]+1]
    else:
        print("Target image dimension is less than the target image dimension")
        return
    
def findGradientMagnitude(imgGrayCircleXDerivative,imgGrayCircleYDerivative):
    return ((imgGrayCircleXDerivative**2+imgGrayCircleYDerivative**2)**(1/2))
```

```python
def findMax2DIndex(resolution):
  location=[0,0]
  maxVal=resolution[0,0]
  for x in range(0,resolution.shape[0]):
    for y in range(0,resolution.shape[1]):
      if(resolution[x,y]>maxVal):
        maxVal=resolution[x,y]
        location=[x,y]
  return location
```

```python
def findMin2DIndex(resolution):
  location=[0,0]
  minVal=resolution[0,0]
  for x in range(0,resolution.shape[0]):
    for y in range(0,resolution.shape[1]):
      if(resolution[x,y]<minVal):
        minVal=resolution[x,y]
        location=[x,y]
  return location
```

```python
def findScaleDown0To255(resolution):
    [xMin,yMin]=findMin2DIndex(resolution)
    [xMax,yMax]=findMax2DIndex(resolution)
    return ((resolution-resolution[xMin,yMin])/(resolution[xMax,yMax]))*255
```

```python
def findNonMaximumSuppression(imgResolutionDirection,imgGrayXYDerivativeMagnitude):
        imgResolutionDirectionNonMaximaSuppressed = np.zeros([imgGrayXYDerivativeMagnitude.shape[0],imgGrayXYDerivativeMagnitude.shape[1]])
        for y in range(1,imgResolutionDirection.shape[0]-1):
            for x in range(1,imgResolutionDirection.shape[1]-1):
                firstNeighbour=0
                secondNeighbour=0
                if (-22.5<=imgResolutionDirection[y,x]<22.5)or(157.5<=imgResolutionDirection[y,x]<180)or(-180<=imgResolutionDirection[y,x]<-157.5):
                    firstNeighbour=imgGrayXYDerivativeMagnitude[y,x-1]
                    secondNeighbour=imgGrayXYDerivativeMagnitude[y,x+1]
                elif (22.5<=imgResolutionDirection[y,x]<67.5)or(-157.5<=imgResolutionDirection[y,x]<=-112.5):
                    firstNeighbour=imgGrayXYDerivativeMagnitude[y+1,x+1]
                    secondNeighbour=imgGrayXYDerivativeMagnitude[y-1,x-1]
                elif (67.5<=imgResolutionDirection[y,x]<112.5)or(-112.5<=imgResolutionDirection[y,x]<=-67.5):
                    firstNeighbour=imgGrayXYDerivativeMagnitude[y+1,x]
                    secondNeighbour=imgGrayXYDerivativeMagnitude[y-1,x]
                elif (112.5<=imgResolutionDirection[y,x]<157.5)or(-67.5<=imgResolutionDirection[y,x]<=-22.5):
                    firstNeighbour=imgGrayXYDerivativeMagnitude[y+1,x-1]
                    secondNeighbour=imgGrayXYDerivativeMagnitude[y-1,x+1]
                if(imgGrayXYDerivativeMagnitude[y,x]>=firstNeighbour)and(imgGrayXYDerivativeMagnitude[y,x]>=secondNeighbour):
                    imgResolutionDirectionNonMaximaSuppressed[y,x]=imgGrayXYDerivativeMagnitude[y,x]
                else:
                    imgResolutionDirectionNonMaximaSuppressed[y,x]=0
        return imgResolutionDirectionNonMaximaSuppressed
```

```python
def findHystereisThresholding(img,threholdHigh,thresholdLow):
    output=np.zeros(img.shape)
    for y in range(0,img.shape[0]):
        for x in range(0,img.shape[1]):
            if(img[y,x]<=thresholdLow):
                output[y,x]=0
            elif(img[y,x]>=threholdHigh):
                output[y,x]=255
            else:
                if(0<x<img.shape[1])and(0<y<img.shape[0]):
                    if(img[y-1:y+2,x-1:x+2].max()>=threholdHigh):
                        output[y,x]=255
                    else:
                        output[y,x]=0
                else:
                    output[y,x]=0
    return output
```

```python
def findCentroid(imgResolution):
    imgCentroidX=0;
    imgCentroidY=0;
    count=0;
    for y in range(0,imgResolution.shape[0]):
        for x in range(0,imgResolution.shape[1]):
            if (imgResolution[y,x]>0):
                count=count+1
                imgCentroidX=imgCentroidX*((count-1)/count)+x/count
                imgCentroidY=imgCentroidY*((count-1)/count)+y/count
    return [int(imgCentroidY),int(imgCentroidX),count]
```

```python
def findEdgeAnglesConsidered(imgDirection,imgResolution,MaxValueAssigned):
        discretizationRange = np.linspace(-180,180,MaxValueAssigned+1)
        print("Considered Angle Boundaries: ", end=" ")
        print(discretizationRange)
        discretizationDirectionSize = np.zeros(discretizationRange.shape[0])
        for y in range(0,imgDirection.shape[0]):
            for x in range(0,imgDirection.shape[1]):
                for index, d in enumerate(discretizationRange,start=0):
                    if ((imgDirection[y,x]<=d)and(imgResolution[y,x]>0)):
                            discretizationDirectionSize[index]=discretizationDirectionSize[index]+1
                            break
        print("Corresponding Angle Frequencies: ", end=" ")
        print(discretizationDirectionSize)
        return discretizationDirectionSize
```

```python
def findRTableAngleDeltaYDeltaX(imgDirection,imgResolution,MaxValueAssigned):
        discretizationRange = np.linspace(-180,180,MaxValueAssigned+1)
        print("Considered Angle Boundaries: ", end=" ")
        print(discretizationRange)
        centroidLocationYX=findCentroid(imgResolution)
        count=0
        AnglesAndEdgesYX = np.zeros([3,imgDirection.shape[0]*imgDirection.shape[1]])
        for y in range(0,imgDirection.shape[0]):
            for x in range(0,imgDirection.shape[1]):
                for index, d in enumerate(discretizationRange,start=0):
                    if ((imgDirection[y,x]<=d)and(imgResolution[y,x]>0)):
                            AnglesAndEdgesYX[0,count]=d
                            AnglesAndEdgesYX[1,count]=y-centroidLocationYX[0]
                            AnglesAndEdgesYX[2,count]=x-centroidLocationYX[1]
                            count=count+1
                            break
        return np.array(list(zip(*sorted(zip(*AnglesAndEdgesYX[0:3,0:count])))))
```

```python
def findAccumulatorArray(imgInterestedShapeEdgesFromRtable, imgTemplateEdgesFromRtable):
    AccumulatorArray=np.zeros([int((imgTemplateEdgesFromRtable[1].max()-imgInterestedShapeEdgesFromRtable[1].min())-(imgTemplateEdgesFromRtable[1].min()-imgInterestedShapeEdgesFromRtable[1].max())),int((imgTemplateEdgesFromRtable[2].max()-imgInterestedShapeEdgesFromRtable[2].min())-(imgTemplateEdgesFromRtable[2].min()-imgInterestedShapeEdgesFromRtable[2].max()))])
    minOffsetY=int((imgTemplateEdgesFromRtable[1].min())-(imgInterestedShapeEdgesFromRtable[1].max()))
    minOffsetX=int((imgTemplateEdgesFromRtable[2].min())-(imgInterestedShapeEdgesFromRtable[2].max()))
    print(minOffsetY)
    print(minOffsetX)
    print(int((imgTemplateEdgesFromRtable[1].max())-(imgInterestedShapeEdgesFromRtable[1].min())))
    print(int((imgTemplateEdgesFromRtable[2].max())-(imgInterestedShapeEdgesFromRtable[2].min())))
    print(imgTemplateEdgesFromRtable[1].max())
    print(AccumulatorArray.shape)
    for shapeIndex, shapeAngle in enumerate(imgInterestedShapeEdgesFromRtable[0],start=0):
        for templateIndex, templateAngle in enumerate(imgTemplateEdgesFromRtable[0],start=0):
            if(shapeAngle==templateAngle):
                y=int(np.absolute(minOffsetY)+(imgTemplateEdgesFromRtable[1,templateIndex]-imgInterestedShapeEdgesFromRtable[1,shapeIndex]))
                x=int(np.absolute(minOffsetX)+(imgTemplateEdgesFromRtable[2,templateIndex]-imgInterestedShapeEdgesFromRtable[2,shapeIndex]))
                if(y<AccumulatorArray.shape[0])and(x<AccumulatorArray.shape[1]):
                    AccumulatorArray[y,x]=AccumulatorArray[y,x]+1
            elif(shapeAngle<templateAngle):
                break
    return np.array(AccumulatorArray)
    
```

```python
def generateRTablefromImg(Img):
    internatl_sigma=0.5
    internatl_T=0.3
    
    [internatl_fx,internatl_fy]=findFilterXYDerivative (internatl_T, internatl_sigma)
    internatl_G=findGaussian(3,internatl_sigma)
    
    internatl_Gx=findConvolution2D(internatl_G,internatl_fx)
    internatl_Gy=findConvolution2D(internatl_G,internatl_fy)
    
    internatl_imgGrayCircleXDerivative = findConvolution2Dfixed(internatl_Gx,Img)
    internatl_imgGrayCircleYDerivative = findConvolution2Dfixed(internatl_Gy,Img)
    
    internatl_imgGrayCircleXYDerivativeMagnitude=findGradientMagnitude(internatl_imgGrayCircleXDerivative,internatl_imgGrayCircleYDerivative)
    
    internatl_imgGrayCircleXYDerivativeMagnitudeScaledDownTo255=findScaleDown0To255(internatl_imgGrayCircleXYDerivativeMagnitude)
    internatl_imgGrayCircleXYDerivativeDirection=findGradientDirection(internatl_imgGrayCircleXDerivative,internatl_imgGrayCircleYDerivative)
    
    internatl_imgGrayCircleXYDerivativeDirectionNMS=findNonMaximumSuppression(internatl_imgGrayCircleXYDerivativeDirection,internatl_imgGrayCircleXYDerivativeMagnitudeScaledDownTo255)
    
    internatl_imgGrayCircleHysteresis=findHystereisThresholding(internatl_imgGrayCircleXYDerivativeDirectionNMS,40,20)
    
    internatl_imgEdgeAngleConsidered=findEdgeAnglesConsidered(internatl_imgGrayCircleXYDerivativeDirection,internatl_imgGrayCircleHysteresis,6)
    internatl_imgEdgeAngleConsideredMatrix=findRTableAngleDeltaYDeltaX(internatl_imgGrayCircleXYDerivativeDirection,internatl_imgGrayCircleHysteresis,6)
    print("I m done")
    #accumulatedArray=findAccumulatorArray(imgEdgeAngleConsideredMatrix,imgEdgeAngleConsideredMatrix+np.array([[0],[0],[0]]))
    
    return internatl_imgEdgeAngleConsideredMatrix
    
```

```python
def findEdgeDominantAngle(imgDirection,imgResolution,MaxValueAssigned):
        discretizationRange = np.linspace(-180,180,MaxValueAssigned+1)
        print("Considered Angle Boundaries: ", end=" ")
        print(discretizationRange)
        discretizationDirectionSize = np.zeros(discretizationRange.shape[0])
        for y in range(0,imgDirection.shape[0]):
            for x in range(0,imgDirection.shape[1]):
                for index, d in enumerate(discretizationRange,start=0):
                    if ((imgDirection[y,x]<=d)and(imgResolution[y,x]>0)):
                            discretizationDirectionSize[index]=discretizationDirectionSize[index]+1
                            break
        print("Corresponding Angle Frequencies: ", end=" ")
        print(discretizationDirectionSize)
        return discretizationRange[np.argmax(discretizationDirectionSize)]
```

```python
def generateRTablefromImgRotationInvariant(Img):
    internatl_sigma=0.5
    internatl_T=0.3
    
    [internatl_fx,internatl_fy]=findFilterXYDerivative (internatl_T, internatl_sigma)
    internatl_G=findGaussian(3,internatl_sigma)
    
    internatl_Gx=findConvolution2D(internatl_G,internatl_fx)
    internatl_Gy=findConvolution2D(internatl_G,internatl_fy)
    
    internatl_imgGrayCircleXDerivative = findConvolution2Dfixed(internatl_Gx,Img)
    internatl_imgGrayCircleYDerivative = findConvolution2Dfixed(internatl_Gy,Img)
    
    internatl_imgGrayCircleXYDerivativeMagnitude=findGradientMagnitude(internatl_imgGrayCircleXDerivative,internatl_imgGrayCircleYDerivative)
    
    internatl_imgGrayCircleXYDerivativeMagnitudeScaledDownTo255=findScaleDown0To255(internatl_imgGrayCircleXYDerivativeMagnitude)
    internatl_imgGrayCircleXYDerivativeDirection=findGradientDirection(internatl_imgGrayCircleXDerivative,internatl_imgGrayCircleYDerivative)
    
    internatl_imgGrayCircleXYDerivativeDirectionNMS=findNonMaximumSuppression(internatl_imgGrayCircleXYDerivativeDirection,internatl_imgGrayCircleXYDerivativeMagnitudeScaledDownTo255)
    
    internatl_imgGrayCircleHysteresis=findHystereisThresholding(internatl_imgGrayCircleXYDerivativeDirectionNMS,40,20)
    
    internatl_imgEdgeAngleConsidered=findEdgeAnglesConsidered(internatl_imgGrayCircleXYDerivativeDirection,internatl_imgGrayCircleHysteresis,6)
    internatl_imgEdgeAngleConsideredMatrix=findRTableAngleDeltaYDeltaX(internatl_imgGrayCircleXYDerivativeDirection,internatl_imgGrayCircleHysteresis,6)
    print("I m done")
    #print(findEdgeDominantAngle(internatl_imgGrayCircleXYDerivativeDirection,internatl_imgGrayCircleHysteresis,6))
    #accumulatedArray=findAccumulatorArray(imgEdgeAngleConsideredMatrix,imgEdgeAngleConsideredMatrix+np.array([[0],[0],[0]]))
    internatl_imgEdgeAngleConsideredMatrix[0]=internatl_imgEdgeAngleConsideredMatrix[0]-findEdgeDominantAngle(internatl_imgGrayCircleXYDerivativeDirection,internatl_imgGrayCircleHysteresis,6)
    return internatl_imgEdgeAngleConsideredMatrix
```
